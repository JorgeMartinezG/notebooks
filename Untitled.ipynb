```python
import json
import pandas as pd
import requests
```

```python
data = requests.get('https://s3-us-west-2.amazonaws.com/fieldcampaigner-data/campaigns/be96d70c75914d749e4250de96a467ef/render/place-of-worship/mapper_engagement.json').json()
user = data[1]
```

```python
user.keys()
```

```python
def create_dict(user):
    # Split contributions per date.
    return [{'user': user['name'], 'type': 'foo', t[0]: t[1]}
                for t in json.loads(user['timeline'])]
users = [create_dict(u) for u in data]
# Flatten list.
users = [item for sublist in users for item in sublist]
```

```python
df = pd.DataFrame(users)
```
